```python
import pandas as pd

df1 = pd.read_csv("../dataco_cleaned.csv")
```

## Delivery Status Distribution

```python
import plotly.express as px

data_delivery_status = df1.groupby(['delivery_status'])['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=False)

fig = px.bar(
    data_delivery_status,
    x='delivery_status',
    y='number_of_orders',
    color='number_of_orders',
    labels={
        'delivery_status': 'Delivery Status',
        'number_of_orders': 'Number of Orders'
    },
    title="Delivery Status Distribution"
)

fig.show()
```

## Delivery Status by Route

```python
import plotly.express as px

data_delivery_status_region = df1.groupby(
    ['delivery_status', 'route']
)['order_id'] \
.count() \
.reset_index(name='number_of_orders') \
.sort_values(by='number_of_orders', ascending=False)

fig = px.bar(
    data_delivery_status_region,
    x='delivery_status',
    y='number_of_orders',
    color='route',
    labels={
        'delivery_status': 'Delivery Status',
        'number_of_orders': 'Number of Orders'
    },
    title="Delivery Status by Route"
)

fig.show()
```

## Top 10 Customers by Orders

```python
import plotly.express as px

# Convert customer_id to string (optional, but fine)
df1['customer_id_str'] = df1['customer_id'].astype(str)

# Group and count orders per customer
data_customers = df1.groupby(['customer_id_str'])['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=False)

# Plot top 10 customers
fig = px.bar(
    data_customers.head(10),
    x='number_of_orders',
    y='customer_id_str',
    color='number_of_orders',
    title="Top 10 Customers by Number of Orders"
    
)

fig.show()
```

## Top 10 Customers by Profit

```python
import plotly.express as px

# Convert customer_id to string (optional)
df1['customer_id_str'] = df1['customer_id'].astype(str)

# Total profit per customer
data_customers_profit = df1.groupby(['customer_id_str'])['profit_per_order'] \
    .sum() \
    .reset_index(name='total_profit') \
    .sort_values(by='total_profit', ascending=False)

# Plot top 10 customers by profit
fig = px.bar(
    data_customers_profit.head(10),
    x='total_profit',
    y='customer_id_str',
    color='total_profit',
    title="Top 10 Customers by Profit"
)

fig.show()
```

## Customer Segment Pie Chart

```python
import plotly.express as px

data_customer_segment = df1.groupby(['customer_segment'])['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=False)

fig = px.pie(
    data_customer_segment,
    values='number_of_orders',
    names='customer_segment',
    title='Number of Orders by Customer Segment',
    width=600,
    height=600,
    color_discrete_sequence=px.colors.sequential.RdBu
)

fig.show()
```

## Orders by Product Category

## Sales Analysis

```python
#Order Country
import plotly.express as px

df_sales_country = df1.groupby(['order_country'])['sales_amount'] \
    .sum() \
    .reset_index(name='total_sales') \
    .sort_values(by='total_sales', ascending=False)

fig = px.bar(
    df_sales_country.head(10),
    x='total_sales',
    y='order_country',
    color='total_sales',
    title="Top 10 Countries by Sales"
)

fig.show()
```

```python
#Product
import plotly.express as px

df_sales_product = df1.groupby(['sku_name'])['sales_amount'] \
    .sum() \
    .reset_index(name='total_sales') \
    .sort_values(by='total_sales', ascending=False)

fig = px.bar(
    df_sales_product.head(10),
    x='total_sales',
    y='sku_name',
    color='total_sales',
    title="Top 10 Products by Sales"
)

fig.show()
```

```python
#Product and deliveray status
import plotly.express as px

df_sales_pd = df1.groupby(['sku_name', 'delivery_status'])['sales_amount'] \
    .sum() \
    .reset_index(name='total_sales') \
    .sort_values(by='total_sales', ascending=False)

fig = px.bar(
    df_sales_pd.head(10),
    x='total_sales',
    y='sku_name',
    color='delivery_status',
    title="Top Products by Sales with Delivery Status"
)

fig.show()
```

```python
#Product and order region

import plotly.express as px

df_sales_pr = df1.groupby(['sku_name', 'route'])['sales_amount'] \
    .sum() \
    .reset_index(name='total_sales') \
    .sort_values(by='total_sales', ascending=False)

fig = px.bar(
    df_sales_pr.head(10),
    x='total_sales',
    y='sku_name',
    color='route',
    title="Top Products by Sales Across Regions"
)

fig.show()
```
