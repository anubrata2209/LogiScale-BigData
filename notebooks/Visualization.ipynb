## Load Path

```python
import pandas as pd
import os

# ── Correct path pointing to project root ──
notebook_dir = os.getcwd()
project_root = os.path.dirname(notebook_dir)

df1 = pd.read_csv(os.path.join(project_root, "dataco_cleaned.csv"))

print(f"Rows loaded : {len(df1):,}")
print(f"Columns     : {df1.columns.tolist()}")
```

## EDA Setup

```python
import pandas as pd
import plotly.express as px
import os

notebook_dir = os.getcwd()
project_root = os.path.dirname(notebook_dir)
output_dir   = os.path.join(project_root, "outputs")

df1 = pd.read_csv(os.path.join(project_root, "dataco_cleaned.csv"))
df1['customer_id_str'] = df1['customer_id'].astype(str)

print(f"Rows    : {len(df1):,}")
print(f"Columns : {len(df1.columns)}")
df1.head(3)
```

## Delivery Status Distribution

```python
data_delivery_status = df1.groupby(['delivery_status'])['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=False)

fig = px.bar(
    data_delivery_status,
    x='delivery_status',
    y='number_of_orders',
    color='number_of_orders',
    color_continuous_scale='Blues',
    labels={
        'delivery_status': 'Delivery Status',
        'number_of_orders': 'Number of Orders'
    },
    title="Delivery Status Distribution"
)
fig.update_layout(plot_bgcolor='white')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_delivery_status.png"))
```

## Delivery Status by Route

```python
data_delivery_status_region = df1.groupby(
    ['delivery_status', 'route']
)['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=False)

fig = px.bar(
    data_delivery_status_region,
    x='delivery_status',
    y='number_of_orders',
    color='route',
    labels={
        'delivery_status': 'Delivery Status',
        'number_of_orders': 'Number of Orders'
    },
    title="Delivery Status by Route"
)
fig.update_layout(plot_bgcolor='white')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_delivery_by_route.png"))
```

## Top 10 Customers by Orders

```python
data_customers = df1.groupby(['customer_id_str'])['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=False)

fig = px.bar(
    data_customers.head(10),
    x='number_of_orders',
    y='customer_id_str',
    color='number_of_orders',
    color_continuous_scale='Teal',
    orientation='h',
    title="Top 10 Customers by Number of Orders",
    labels={
        'customer_id_str': 'Customer ID',
        'number_of_orders': 'Number of Orders'
    }
)
fig.update_layout(plot_bgcolor='white', yaxis={'categoryorder': 'total ascending'})
fig.show()
fig.write_image(os.path.join(output_dir, "chart_top10_customers_orders.png"))
```

## Top 10 Customers by Profit

```python
data_customers_profit = df1.groupby(['customer_id_str'])['profit_per_order'] \
    .sum() \
    .reset_index(name='total_profit') \
    .sort_values(by='total_profit', ascending=False)

fig = px.bar(
    data_customers_profit.head(10),
    x='total_profit',
    y='customer_id_str',
    color='total_profit',
    color_continuous_scale='Greens',
    orientation='h',
    title="Top 10 Customers by Total Profit",
    labels={
        'customer_id_str': 'Customer ID',
        'total_profit': 'Total Profit ($)'
    }
)
fig.update_layout(plot_bgcolor='white', yaxis={'categoryorder': 'total ascending'})
fig.show()
fig.write_image(os.path.join(output_dir, "chart_top10_customers_profit.png"))
```

## Customer Segment Pie Chart

```python
data_customer_segment = df1.groupby(['customer_segment'])['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=False)

fig = px.pie(
    data_customer_segment,
    values='number_of_orders',
    names='customer_segment',
    title='Number of Orders by Customer Segment',
    width=600,
    height=600,
    color_discrete_sequence=px.colors.sequential.RdBu,
    hole=0.3   # makes it a donut chart — looks more modern
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_customer_segment.png"))
```

## Orders by Product Category

```python
data_category = df1.groupby(['category'])['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=True)

fig = px.bar(
    data_category,
    x='number_of_orders',
    y='category',
    color='number_of_orders',
    color_continuous_scale='Purples',
    orientation='h',
    title="Orders by Product Category",
    labels={
        'category': 'Category',
        'number_of_orders': 'Number of Orders'
    }
)
fig.update_layout(plot_bgcolor='white', height=600)
fig.show()
fig.write_image(os.path.join(output_dir, "chart_orders_by_category.png"))
```

##  Orders by Region + Top 10 Countries

```python
# Orders by Region
data_region = df1.groupby(['route'])['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=True)

fig = px.bar(
    data_region,
    x='number_of_orders',
    y='route',
    color='number_of_orders',
    color_continuous_scale='Oranges',
    orientation='h',
    title="Orders by Region",
    labels={'route': 'Region', 'number_of_orders': 'Number of Orders'}
)
fig.update_layout(plot_bgcolor='white')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_orders_by_region.png"))

# Top 10 Countries
data_countries = df1.groupby(['order_country'])['order_id'] \
    .count() \
    .reset_index(name='number_of_orders') \
    .sort_values(by='number_of_orders', ascending=True)

fig = px.bar(
    data_countries.head(10),
    x='number_of_orders',
    y='order_country',
    color='number_of_orders',
    color_continuous_scale='Reds',
    orientation='h',
    title="Top 10 Countries by Orders",
    labels={'order_country': 'Country', 'number_of_orders': 'Number of Orders'}
)
fig.update_layout(plot_bgcolor='white')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_top10_countries_orders.png"))
```

## Sales Analysis (All 5 Charts)

##  1. Top 10 Countries by Sales

```python
df_sales_country = df1.groupby(['order_country'])['sales_amount'] \
    .sum().reset_index(name='total_sales') \
    .sort_values(by='total_sales', ascending=True)

fig = px.bar(df_sales_country.head(10), x='total_sales', y='order_country',
    color='total_sales', color_continuous_scale='Blues', orientation='h',
    title="Top 10 Countries by Sales",
    labels={'order_country': 'Country', 'total_sales': 'Total Sales ($)'})
fig.update_layout(plot_bgcolor='white')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_sales_by_country.png"))
```

##  2. Top 10 Products by Sales

```python
df_sales_product = df1.groupby(['sku_name'])['sales_amount'] \
    .sum().reset_index(name='total_sales') \
    .sort_values(by='total_sales', ascending=True)

fig = px.bar(df_sales_product.head(10), x='total_sales', y='sku_name',
    color='total_sales', color_continuous_scale='Teal', orientation='h',
    title="Top 10 Products by Sales",
    labels={'sku_name': 'Product', 'total_sales': 'Total Sales ($)'})
fig.update_layout(plot_bgcolor='white')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_sales_by_product.png"))
```

## 3. Top Products by Sales with Delivery Status

```python
df_sales_pd = df1.groupby(['sku_name', 'delivery_status'])['sales_amount'] \
    .sum().reset_index(name='total_sales') \
    .sort_values(by='total_sales', ascending=False)

fig = px.bar(df_sales_pd.head(10), x='total_sales', y='sku_name',
    color='delivery_status', orientation='h',
    title="Top Products by Sales with Delivery Status",
    labels={'sku_name': 'Product', 'total_sales': 'Total Sales ($)'})
fig.update_layout(plot_bgcolor='white')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_sales_product_delivery.png"))
```

## 4. Top Products by Sales Across Regions

```python
df_sales_pr = df1.groupby(['sku_name', 'route'])['sales_amount'] \
    .sum().reset_index(name='total_sales') \
    .sort_values(by='total_sales', ascending=False)

fig = px.bar(df_sales_pr.head(10), x='total_sales', y='sku_name',
    color='route', orientation='h',
    title="Top Products by Sales Across Regions",
    labels={'sku_name': 'Product', 'total_sales': 'Total Sales ($)'})
fig.update_layout(plot_bgcolor='white')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_sales_product_region.png"))
```

## 5. Top Categories by Sales

```python
df_sales_category = df1.groupby(['category'])['sales_amount'] \
    .sum().reset_index(name='total_sales') \
    .sort_values(by='total_sales', ascending=True)

fig = px.bar(df_sales_category.head(10), x='total_sales', y='category',
    color='total_sales', color_continuous_scale='Greens', orientation='h',
    title="Top Categories by Sales",
    labels={'category': 'Category', 'total_sales': 'Total Sales ($)'})
fig.update_layout(plot_bgcolor='white')
fig.show()
fig.write_image(os.path.join(output_dir, "chart_sales_by_category.png"))

print(f"\n  All EDA charts saved to: {output_dir}")
```
